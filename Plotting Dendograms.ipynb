**IMPORTING THE LIBRARIES**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**STORING GIVEN POINTS NP_ARRAY X**

```python
X = np.array([[1,1],[4,1],[5,2],[6,2],[2,3],])
```

**LABELING EACH POINT AND PLOTTING A SCATTER PLOT**

```python
labels = range(1,6)
plt.figure(figsize=(12, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

**DENDROGRAM FOR AGGLOMERATIVE CLUSTERING**

**CONSIDERING LINKAGE IS SINGLE**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
cluster.fit_predict(X)
linked = linkage(X,"single")
labelList = range(1,6)
plt.figure(figsize=(10, 7))
dendrogram(linked,
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True,above_threshold_color='#b329dd',orientation='right')
plt.show()
```

**CONSIDERING LINKAGE IS COMPLETE**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
cluster.fit_predict(X)
linked = linkage(X,"complete")
labelList = range(1,6)
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='right',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True,above_threshold_color='#b329dd')
plt.show()
```

**CONSIDERING LINKAGE IS AVERAGE**

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')
cluster.fit_predict(X)
linked = linkage(X,"average")
labelList = range(1,6)
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='right',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True,above_threshold_color='#b329dd')
plt.show()
```
