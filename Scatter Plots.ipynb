**IMPORTING THE LIBRARIES**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**STORING GIVEN POINTS IN NP_ARRAY AS X**

```python
X = np.array([[1,1],[4,1],[5,2],[6,2],[2,3],])
```

**LABELING EACH POINT AND PLOTTING A SCATTER PLOT**

```python
labels = range(1,6)
plt.figure(figsize=(12, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

**IMPLEMENTATION OF AGGLOMERATIVE CLUSTERING**

**CONSIDERING LINKAGE IS SINGLE**

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
cluster.fit_predict(X)
print(cluster.labels_)
colors=['black','yellow']
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap=matplotlib.colors.ListedColormap(colors))
```

**CONSIDERING LINKAGE IS AVERAGE**

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')
cluster.fit_predict(X)
print(cluster.labels_)
colors=['yellow','black']
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap=matplotlib.colors.ListedColormap(colors))
```

**CONSIDERING LINKAGE IS COMPLETE**

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
cluster.fit_predict(X)
print(cluster.labels_)
colors=['yellow','black']
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap=matplotlib.colors.ListedColormap(colors))
```

**INFERENCE : WE SEE THAT THE AGNES CLUSTERING ALGORITHM WORKS IN TOP-DOWN APPROACH AND CLUSTERS THE SAME DATA SET INTO TWO GROUPS UNIQUELY FOR SINGLE,AVERAGE AND COMPLETE LINKAGES.**
